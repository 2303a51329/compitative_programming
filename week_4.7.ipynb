```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)  # 1-based indexing

    # Add delta at index
    def update(self, index, delta):
        while index <= self.n:
            self.tree[index] += delta
            index += index & -index

    # Prefix sum from 1 to index
    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total


def main():
    # =========================
    # TASK 1: Student Attendance
    # =========================
    print("---- Student Attendance Analysis ----")

    attendance = [1, 2, 1, 2, 1]
    n1 = len(attendance)

    attendanceFT = FenwickTree(n1)

    # Build Fenwick Tree
    for i in range(n1):
        attendanceFT.update(i + 1, attendance[i])

    # Query: total till Day 4
    print("Total attendance till Day 4 =", attendanceFT.query(4))

    # Update Day 3 from 1 → 2
    oldAttendance = attendance[2]
    newAttendance = 2
    delta1 = newAttendance - oldAttendance

    attendanceFT.update(3, delta1)
    attendance[2] = newAttendance

    print("After update, total attendance till Day 4 =", attendanceFT.query(4))

    # =========================
    # TASK 2: Daily Sales Revenue
    # =========================
    print("\n---- Daily Sales Revenue Analysis ----")

    revenue = [200, 150, 300, 250, 100, 180, 220]
    n2 = len(revenue)

    revenueFT = FenwickTree(n2)

    # Build Fenwick Tree
    for i in range(n2):
        revenueFT.update(i + 1, revenue[i])

    # Query: total till Day 6
    print("Total sales revenue till Day 6 =", revenueFT.query(6))

    # Update Day 5 from 100 → 160
    oldRevenue = revenue[4]
    newRevenue = 160
    delta2 = newRevenue - oldRevenue

    revenueFT.update(5, delta2)
    revenue[4] = newRevenue

    print("After update, total sales revenue till Day 6 =", revenueFT.query(6))


if __name__ == "__main__":
    main()
```
