```python
def max_crossing_sum(arr, left, mid, right):
    # Max suffix sum on left side
    left_sum = float('-inf')
    total = 0
    for i in range(mid, left - 1, -1):
        total += arr[i]
        left_sum = max(left_sum, total)

    # Max prefix sum on right side
    right_sum = float('-inf')
    total = 0
    for i in range(mid + 1, right + 1):
        total += arr[i]
        right_sum = max(right_sum, total)

    return left_sum + right_sum


def max_subarray_sum(arr, left, right):
    # Base case: one element
    if left == right:
        return arr[left]

    mid = (left + right) // 2

    # Recursive cases
    left_max = max_subarray_sum(arr, left, mid)
    right_max = max_subarray_sum(arr, mid + 1, right)
    cross_max = max_crossing_sum(arr, left, mid, right)

    return max(left_max, right_max, cross_max)


# Driver code
T = int(input())
for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))
    print(max_subarray_sum(arr, 0, N - 1))
```
