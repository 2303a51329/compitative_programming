```python
def max_meetings(meetings):
    meetings.sort(key=lambda x: x[1])

    count = 0
    last_end = -1
    for start, end in meetings:
        if start >= last_end:
            count += 1
            last_end = end

    return count
T = int(input())
for _ in range(T):
    N = int(input())
    meetings = []
    for _ in range(N):
        Si, Ei = map(int, input().split())
        meetings.append((Si, Ei))

    print(max_meetings(meetings))
```
